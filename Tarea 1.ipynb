# Características del país y su relación con el GDP per Cápita
* PopulationArea (sq. mi.):
No existe una relación directa con el GDP per cápita, pero países con grandes extensiones pueden tener economías más diversificadas y recursos naturales abundantes, sin embargo, si el area es grande y la población es baja, podria haber costos más altos en infraestructuras y servicios.

* Pop. Density (per sq. mi.):
La densidad de población hace referencia a la cantidad de personas que viven en un determinado territorio (habitantes por km cuadrado), una alta densidad puede indicar economias mas productivas y urbanizadas, en cambio, una densidad más baja, podria sugerir economias con menor desarrollo industrial.

* Coastline (coast/area ratio): Paises con mas costas, tienen mayor acceso a puertos, lo que impularía el crecimiento economico.

* Net migration:
Esta variable hace referencia al el saldo migratorio,el cual tiene una correlación positiva, es decir, más inmigrantes que emigrantes, sugiere una economia más atractiva, con oportunidades laborales, en cambio, si el saldo migratorio es negativo, podría ser un signo de inestabilidad economica.

* Infant mortality (per 1000 births):
Generalmente, un menor indice de mortalidad infantil, indica un mejor sistema de salud, por ende una mejor economía, estando así, correlacionada negativamente con el GDP per cápita.

* Literacy (%):
Una mayor tasa de alfebitación suele estar relacionada con una economia más productiva, ya que permite una fuerza laboral más calificada.

* Phones (per 1000):
Un mayor acceso a tecnologias de la comunicación, sugiere un mayor desarrollo economico, debido a que una mejor conectividad mejora la calidad de la educación, el comercio y la productividad de los habitantes del pais.


* Arable (%):
Si existe mayor porcentaje de tierras arables en un determinado pais, esto podria significar una economía más agricola, sin embargo, no implica que haya un mayor GDP per cápita.

* Crops (%):
Si existe mayor porcentaje de cultivos en un determinado pais, esto podria significar una economía en desarrollo.

* Other (%):
Representa el porcentaje de tierra que no es ni agricola ni de cultivos, lo que puede incluir otras zonas. por ejemplo: áreas urbanizadas, montañas, bosques, reservas naturales), el impacto en el GDP per capíta dependerá de que tan aprovechadas sean estas tierras.

* Climate:
Variable que hace referencia al clima del pais, por ejemplo, climas templados suelen ser más favorables para la productividad economica.

* Birthrate:
Un indice de natalidad bajo suele darse en paises con una economía avanzada, con mayor GDP percapíta, mientras que, un alto indice de natalidad puede estar relacionado con econommias en desarrollo, donde el GDP suele ser más bajo.

* Deathrate:
Un pais con tasas de mortalidad bajo, sugieren un mayor GDP, debido a mejores servicios de salud.

* Agriculture:
Economias basada en la agricultura, suelen tener menor GDP per cápita.

* Industry:
Un pais con fuerte base industrial sugiere un mayor GDP per cápita.

* Service:
Las economias basadas en servicios suelen tener altos niveles de GDP, ya que implican mayor especialización y tecnología.

```python
#Dropeamos la obs 223 correspondiente a 'Western Sahara' con GDP per capita NaN
df=df.drop(223, axis=0)
#Verificamos NaN nuevamente
numero_nan2 = df.isna().sum()
numero_nan2
```

```python
#Generamos un Loop para cada variable


for var in df.columns.values:
  #Si el numero de Nan de la variable es cero pasamos a la siguiente variable
    if df[var].isnull().sum() == 0:
        continue
  #Para la variable Clima, como es discreta, para imputar los Nan utilizamos la moda (de acuerdo a practicas estandar)
    if var == 'Climate':
        Valores_Imputados = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
  #Para variables continuas, para imputar los Nan utilizamos la mediana (de acuerdo a practicas estandar)
    else:
        Valores_Imputados = df.groupby('Region')[var].median()
  #Para cada variable, remplazamos la mediana/moda calculada por región
    for region in df['Region'].unique():
        df.loc[(df[var].isnull())&(df['Region']==region), var] = Valores_Imputados[region]

#Verificamos si quedan Nan
numero_nan = df.isna().sum()
numero_nan
```




---
Realizamos un conteo de los missing values para cada una de las variables en estudio, para tener un mejor contexto que nos permitiera el correcto método de imputación para missing values.

Mediante el conteo, se detectó que la Variable GDP per Capita presentaba un solo Missing Value. Dado que es la variable de estudio principal, se decidió eliminar la observación con el missing value que corresponde a "Western Sahara". No obtenemos beneficio al imputar algún valor al missing value porque no aporta informacion o cambios a las conclusiones de las variables estudiadas, en principa el GDP per capita.

Para que los datos mantengan coherencia con lo esperado, se decidió imputar los valores missings reemplazandolos con el valor de la mediana/moda calculada por Región.

Para el caso de la variable Climate, dado que es una variable discreta, se decidió realizar la imputación mediante el reemplazo de missing values con la moda calculada por región.

Para el resto de variables, dado que son continuas, se decidió realizar la imputación mediante el reemplazo de missing values con la mediana calculada por región. Se dedició utilizar la mediana en lugar de la media, para reducir el probable efecto de valores outliers.

### Pregunta 1.15

En los ejemplos anteriores calculamos correlaciones para `GDP ($ per capita)` y `GDP ($)`. Genere un nuevo dataframe que tenga le variación porcentual de la correlación absoluta para cada una de las columnas de características, e.g., si la correlación en valor absoluto de `GDP ($ per capita)` vs `Industry` es 0.1 y la correlación `GDP ($)` vs `Industry` es 0.5, la variación deberá ser +500%. Dicha variación porcentual puede ser positiva o negativa, pero ordene los el dataframe de tal manera que la variación de correlación absoluta sea desendiente.

```python
#Dropeamos datos no numericos y los GDPs y asiganmos a nuevo DF segun GDP o GDP per capita
df_gdp_pc=df.drop(['date','Country','Region','GDP($)'],axis=1)
df_gdp=df.drop(['date','Country','Region','GDP ($ per capita)'],axis=1)

#Calculamos las correlaciones en valor absoluto 
df_corr_abs = pd.DataFrame ()
df_corr_abs["Corr Abs con GDP ($ per capita)"]=abs(df_gdp_pc.corrwith(df["GDP ($ per capita)"]))
df_corr_abs["Corr Abs con GDP($)"]=abs(df_gdp.corrwith(df["GDP($)"]))

#Dropeamos GDP per capita pues no es de interes para el analisis
df_corr_abs=df_corr_abs.drop(["GDP ($ per capita)"], axis=0)

#Calculamos Variacion porcentual de Correlaciones Absolutas
df_corr_abs["Variación Porcentual de Corr Abs"]= 100*((df_corr_abs["Corr Abs con GDP($)"]-df_corr_abs["Corr Abs con GDP ($ per capita)"])/df_corr_abs["Corr Abs con GDP ($ per capita)"])
df_corr_abs["Variacion en dec"]= ((df_corr_abs["Corr Abs con GDP($)"]-df_corr_abs["Corr Abs con GDP ($ per capita)"])/df_corr_abs["Corr Abs con GDP ($ per capita)"])


#Ordenamos descendentemente 
df_corr_abs = df_corr_abs.sort_values(by="Variación Porcentual de Corr Abs", ascending=False)

df_corr_abs.head(30)
```

Es decir, asigne un número de 1 a 5 a $\{$ `Solid Fuel`, `Liquid Fuel`, `Gas Fuel`, `Cement`, `Gas Flaring` $\}$, donde 1 es la mayor fuente de emisión de ese país en aquel año, y 5 indica que fue la menor; así para todos los países.

Si en 2007 no se reporta una fuente de emisión para un país, por ejemplo, si emisiones de `Gas Flaring` no se reportara, entonces asigne números de 1 a 4 a las fuentes restantes. Análogo para un menor número de datos.

Luego, por cada variable grafique un histograma de frecuencias del ranking que obtuvo la fuente emisión a lo largo de todos los países.

¿Cuál fue la fuente más contaminante en la mayoría de países en 2007?


```python
#definimod funcion que genere ranking
def ranking_emisiones_co2(df):
    #mantenemod datos solo de 2007
    df_2007 = df[df['Year'] == '2007-01-01'].copy()  
    #definimos variables parte del ranking
    fuentes = ['Solid Fuel', 'Liquid Fuel', 'Gas Fuel', 'Cement', 'Gas Flaring']

#Definimos función 
    def rank_var(row):
        #Verifica que las variables de emisión no sean Nan o Cero
        emisiones = {var: row[var] for var in fuentes if pd.notna(row[var]) and row[var] != 0}
        #Ordena variables de emisiones de pais de forma descendete de acuedo a sus valores
        var_ordenadas = sorted(emisiones, key=emisiones.get, reverse=True)
        #Genera ranking de acuerdo al orden de la variable (por su valor)
        ranking = {var: rank + 1 for rank, var in enumerate(var_ordenadas)}
        #Genera nuevas variables con el puesto del ranking con valores de 1 a 5
        for var in fuentes:
            row[f'Rank_{var}'] = ranking.get(var)
        return row

#generamos df_ranke aplicando función rank var a los datos de 2007
    df_ranked = df_2007.apply(rank_var, axis=1)

    # Ordenar las columnas para mejor visualización
    cols_to_order = ['Country'] + [col for col in df_ranked if col.startswith('Rank')]
    df_ranked = df_ranked[cols_to_order]
    return df_ranked

# Aplicamos función de ranking a df_co2 y exportamos a excel para verificar resultados
df_ranked_co2 = ranking_emisiones_co2(df_co2)
df_ranked_co2.to_excel('output2.xlsx')
```

En otras palabras, genere una serie con _**desvíos del cambio porcentual promedio**_ $\forall i,t$:

```python
def calcular_media(df, columna_valor, columna_pais, columna_fecha, nueva_columna_nombre):
    df = df.sort_values(by=[columna_pais, columna_fecha])
    df[nueva_columna_nombre] = df.groupby(columna_pais)[columna_valor].transform('mean')
    return df

df_co2 = calcular_media(df_co2, 'Total', 'Country', 'Year', 'Media')

df_co2["NuevaSerie"] = np.where(df_co2["Variacion_Porcentual"].notna(), df_co2["Variacion_Porcentual"] - df_co2["Media"], np.nan)
df_co2.to_excel('output180.xlsx')
```
