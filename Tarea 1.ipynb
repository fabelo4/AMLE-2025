```python
#Hallamos las variables con Nan y el numero de Nan
numero_nan = df.isna().sum()
numero_nan
```

```python
#Para la variable GDP, se decidió dropear el unico valor Missing
#identificamos el numero de la obs
df.loc[df['GDP ($ per capita)'].isnull()]
```

```python
#Dropeamos la obs 223 correspondiente a 'Western Sahara' con GDP per capita NaN
df=df.drop(223, axis=0)
#Verificamos NaN nuevamente
numero_nan2 = df.isna().sum()
numero_nan2
```

```python
#Generamos un Loop para cada variable


for var in df.columns.values:
  #Si el numero de Nan de la variable es cero pasamos a la siguiente variable
    if df[var].isnull().sum() == 0:
        continue
  #Para la variable Clima, como es discreta, para imputar los Nan utilizamos la moda (de acuerdo a practicas estandar)
    if var == 'Climate':
        Valores_Imputados = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
  #Para variables continuas, para imputar los Nan utilizamos la mediana (de acuerdo a practicas estandar)
    else:
        Valores_Imputados = df.groupby('Region')[var].median()
  #Para cada variable, remplazamos la mediana/moda calculada por región
    for region in df['Region'].unique():
        df.loc[(df[var].isnull())&(df['Region']==region), var] = Valores_Imputados[region]

#Verificamos si quedan Nan
numero_nan = df.isna().sum()
numero_nan
```




---
Realizamos un conteo de los missing values para cada una de las variables en estudio, para tener un mejor contexto que nos permitiera el correcto método de imputación para missing values.

Mediante el conteo, se detectó que la Variable GDP per Capita presentaba un solo Missing Value. Dado que es la variable de estudio principal, se decidió eliminar la observación con el missing value que corresponde a "Western Sahara".

Para que los datos mantengan coherencia con lo esperado, se decidió imputar los valores missings reemplazando los con el valor de la mediana/moda calculada por Región.

Para el caso de la variable Climate, dado que es una variable discreta, se decidió realizar la imputación mediante el reemplazo de missing values con la moda calculada por región.

Para el resto de variables, dado que son continuas, se decidió realizar la imputación mediante el reemplazo de missing values con la mediana calculada por región. Se dedició utilizar la mediana en lugar de la media, para reducir el probable efecto de valores outliers.

### Pregunta 1.15

En los ejemplos anteriores calculamos correlaciones para `GDP ($ per capita)` y `GDP ($)`. Genere un nuevo dataframe que tenga le variación porcentual de la correlación absoluta para cada una de las columnas de características, e.g., si la correlación en valor absoluto de `GDP ($ per capita)` vs `Industry` es 0.1 y la correlación `GDP ($)` vs `Industry` es 0.5, la variación deberá ser +500%. Dicha variación porcentual puede ser positiva o negativa, pero ordene los el dataframe de tal manera que la variación de correlación absoluta sea desendiente.

```python
df_gdp_pc=df.drop(['date','Country','Region','GDP($)'],axis=1)
df_gdp=df.drop(['date','Country','Region','GDP ($ per capita)'],axis=1)

df_corr_abs = pd.DataFrame ()
df_corr_abs["Corr Abs con GDP ($ per capita)"]=abs(df_gdp_pc.corrwith(df["GDP ($ per capita)"]))
df_corr_abs["Corr Abs con GDP($)"]=abs(df_gdp.corrwith(df["GDP($)"]))

df_corr_abs=df_corr_abs.drop(["GDP ($ per capita)"], axis=0)

df_corr_abs["Variación Porcentual de Corr Abs"]= (df_corr_abs["Corr Abs con GDP($)"]/df_corr_abs["Corr Abs con GDP ($ per capita)"])
df_corr_abs["Variacion en dec"]= (df_corr_abs["Corr Abs con GDP($)"]/df_corr_abs["Corr Abs con GDP ($ per capita)"])

df_corr_abs.loc[df_corr_abs["Variación Porcentual de Corr Abs"] > 1, "Variación Porcentual de Corr Abs"] *= 100
df_corr_abs.loc[df_corr_abs["Variación Porcentual de Corr Abs"] <= 1, "Variación Porcentual de Corr Abs"] *= -100

df_corr_abs = df_corr_abs.sort_values(by="Variación Porcentual de Corr Abs", ascending=False)

df_corr_abs.head(30)
```
